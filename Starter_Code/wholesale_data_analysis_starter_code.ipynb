## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["subtotal"] = df["unit_price"] * df["qty"]
subtotal_df = df[["unit_price", "qty", "subtotal"]]
subtotal_df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def price(total_weight):
    if total_weight > 50:
        return 7
    else:
        return 10

    
df["shipping_price"] = df.apply(lambda row: row["unit_weight"] * row["qty"] * price(row["unit_weight"] * row["qty"]), axis=1)
shipping_price_df = df[["unit_price", "unit_weight", "qty",	"shipping_price"]]
shipping_price_df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["total_price"] = df.apply(lambda row: ((row["subtotal"] + row["shipping_price"]) * .0925 + row["subtotal"] + row["shipping_price"]), axis=1).round(2)
total_price_df = df[["subtotal", "shipping_price", "total_price"]]
total_price_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["total_line_cost"] = df.apply(lambda row: ((row["unit_cost"] * row["qty"]) + row["shipping_price"]), axis=1)
total_line_cost_df = df[["unit_cost", "qty", "shipping_price", "total_line_cost"]]
total_line_cost_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df["line_profit"] = df.apply(lambda row: (row["total_price"] - row["total_line_cost"]), axis = 1)
profit_df = df[["total_price", "total_line_cost", "line_profit"]]
profit_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

top5_df = df.groupby('client_id')['qty'].sum().reset_index().sort_values('qty', ascending=False).head(5)
top5_df
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
columns_to_sum = ['qty', 'shipping_price', 'total_price', 'line_profit']
top5_summary_df = df.groupby('client_id')[columns_to_sum].sum().reset_index().sort_values('qty', ascending=False).head(5)
top5_summary_df = top5_summary_df.reset_index(drop=True)
top5_summary_df.rename(columns={"qty": "Total Units Purchased", "shipping_price" : "Total Shipping Price",
                       "total_price" : "Total Revenue", "line_profit" : "Total Profit", "client_id" : "Client_ID"},
                         inplace=True)
top5_summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.


# Define the money columns. 
money_columns = ["Total Shipping Price", "Total Revenue", "Total Profit"]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(amount):
    millions = amount / 1000000
    return millions

# Apply the currency_format_millions function to only the money columns. 
top5_summary_df[money_columns] = top5_summary_df[money_columns].apply(currency_format_millions, axis=1)

# Rename the columns to reflect the change in the money format. 
top5_summary_df.rename(columns={"Total Shipping Price" : "Total Shipping Price (millions)",
                                "Total Revenue" : "Total Revenue (millions)", 
                                "Total Profit" : "Total Profit (millions)"},
                                inplace=True)

top5_summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
final_df = top5_summary_df.sort_values(by= "Total Profit (millions)", ascending=False)
final_df
```
